# Lab 12 Class Inheritance

ในการทดลองนี้ จะได้ศึกษาลำดับของการทำงานของคลาสในการสืบทอดสมบัติ

โดยให้คอนสตรักเตอร์ของคลาสแต่ละระดับทำหน้าที่แสดงข้อความออกมาตามบลำดับการสร้าง object


## แผนผังการสืบทอดสมบัติของคลาส

````
classDiagram
    class BaseClass1
    class BaseClass2
    class DerivedClass

    BaseClass1 <|-- DerivedClass
    
````

จากแผนภาพคลาสไดอะแกรมด้านบน เราสามารถนำมาเขียนโปรแกรมในภาษา C# ได้ดังนี้

1. สร้าง console application project

```cmd
dotnet new console --name Lab12_Ex03
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
DerivedClass dc = new DerivedClass();
class BaseClass1
 {
    public BaseClass1()
    {
        System.Console.WriteLine("This is BaseClass1");
    }
 }
class BaseClass2
 {
    public BaseClass2()
    {
        System.Console.WriteLine("This is BaseClass2");
    }
 }

class DerivedClass : BaseClass1
{
    public DerivedClass()
    {
        System.Console.WriteLine("This is DerivedClass");
    }
}
```



3.Build project โดยการใช้คำสั่ง

```cmd
dotnet build  Lab12_Ex03
```

ถ้ามีที่ผิดพลาดในโปรแกรม ให้แก้ไขให้ถูกต้อง



4.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 3



5.Run project โดยการใช้คำสั่ง

```cmd
dotnet run --project Lab12_Ex03
```



$ dotnet run --project Lab12_Ex03
This is BaseClass1
This is DerivedClass

6.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 5



7.อธิบายสิ่งที่พบในการทดลอง

BaseClass1 บรรทัดแรกของผลลัพธ์ "This is BaseClass1" แสดงว่า Constructor ของคลาส BaseClass1 ถูกเรียกใช้งานก่อน
DerivedClass บรรทัดที่สอง "This is DerivedClass" แสดงว่า Constructor ของคลาส DerivedClass ถูกเรียกใช้งานหลังจาก BaseClass1

## คำถาม

1. การที่มีคลาส BaseClass2 อยู่ใน source code นั้นมีผลใดๆ ต่อการทำงานของโปรแกรมหรือไม่

__ตอบ__

การที่มีคลาส BaseClass2 อยู่ใน source code นั้นไม่มีผลใดๆ ต่อการทำงานของโปรแกรม เนื่องจากไม่มีการประกาศใช้งานหรือเรียกใช้คลาส BaseClass2 ในโค้ดส่วนอื่นๆ เลย ดังนั้น คลาส BaseClass2 จึงเป็นเพียงคลาสที่ถูกกำหนดไว้เฉยๆ แต่ไม่ได้มีส่วนร่วมในการทำงานของโปรแกรม

2. ให้บอกลำดับการทำงานของคลาส ว่า constructor มีลำดับการเรียกอย่างไร (เรียก base class หรือ derive class ก่่อน)

__ตอบ__

ลำดับการทำงานของ constructor คือ เรียก constructor ของ base class ก่อน แล้วจึงเรียก constructor ของ derived class
