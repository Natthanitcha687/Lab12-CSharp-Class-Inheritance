# Lab 12 Class Inheritance

## Multiple base class inheritance

````
classDiagram
    class BaseClass1
    class BaseClass2
    class DerivedClass

    BaseClass1 <|-- BaseClass2
    BaseClass2 <|-- DerivedClass
````

1. สร้าง console application project

```cmd
dotnet new console --name Lab12_Ex06
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
DerivedClass dc = new DerivedClass();
class BaseClass1
 {
    public BaseClass1()
    {
        System.Console.WriteLine("This is BaseClass1");
    }
 }
class BaseClass2: BaseClass1
 {
    public BaseClass2()
    {
        System.Console.WriteLine("This is BaseClass2");
    }
 }

class DerivedClass : BaseClass2
{
    public DerivedClass()
    {
        System.Console.WriteLine("This is DerivedClass");
    }
}
```



3.Build project โดยการใช้คำสั่ง

```cmd
dotnet build  Lab12_Ex06
```

ถ้ามีที่ผิดพลาดในโปรแกรม ให้แก้ไขให้ถูกต้อง



4.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 3



5.Run project โดยการใช้คำสั่ง

```cmd
dotnet run --project Lab12_Ex06
```



6.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 5



7.อธิบายสิ่งที่พบในการทดลอง

BaseClass1: บรรทัดแรกของผลลัพธ์ "This is BaseClass1" แสดงว่า Constructor ของคลาส BaseClass1 ถูกเรียกใช้งานก่อน

BaseClass2: บรรทัดที่สอง "This is BaseClass2" แสดงว่า Constructor ของคลาส BaseClass2 ถูกเรียกใช้งานหลังจาก BaseClass1

DerivedClass: บรรทัดที่สาม "This is DerivedClass" แสดงว่า Constructor ของคลาส DerivedClass ถูกเรียกใช้งานเป็นลำดับสุดท้าย
